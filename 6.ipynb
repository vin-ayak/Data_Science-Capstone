```python
import folium
from folium.plugins import MarkerCluster

# Create a new column 'marker_color' based on the 'class' value
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Loop through the launch records and add markers with the appropriate color to the MarkerCluster
for index, row in spacex_df.iterrows():
    # Create a marker with the specified color based on 'marker_color' column
    marker = folium.Marker(
        location=[row['Lat'], row['Long']],
        icon=folium.Icon(color=row['marker_color']),
        popup=row['Launch Site'] + (" (Success)" if row['class'] == 1 else " (Failure)")
    )
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Create a folium map centered at a location (e.g., NASA Johnson Space Center)
site_map = folium.Map(location=[29.559684888503615, -95.0830971930759], zoom_start=5)

# Add the MarkerCluster to the map
site_map.add_child(marker_cluster)

# Display the map
site_map
```

```python
import folium
from folium.plugins import MarkerCluster

# Create a MarkerCluster object
marker_cluster = MarkerCluster()

# Loop through the launch records and add markers to the MarkerCluster
for index, record in spacex_df.iterrows():
    # Determine the marker color based on the 'class' value (0 for failure, 1 for success)
    marker_color = 'green' if record['class'] == 1 else 'red'
    
    # Create a folium.Marker for the launch record
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],  # Launch site coordinates
        icon=folium.Icon(color=marker_color),      # Marker color
        popup=f"Launch Site: {record['Launch Site']}\nOutcome: {'Success' if record['class'] == 1 else 'Failure'}"
    )
    
    # Add the marker to the MarkerCluster
    marker_cluster.add_child(marker)

# Create a folium map centered at a location (e.g., NASA Johnson Space Center)
site_map = folium.Map(location=[29.559684888503615, -95.0830971930759], zoom_start=4)

# Add the MarkerCluster to the map
site_map.add_child(marker_cluster)

# Display the map
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
